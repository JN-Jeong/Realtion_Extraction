#### ver1: subject랑 object entity 바꾼 데이터 추가

```python
i = 2
# 1. 원본 문장 확인
sentence = train_ver2['sentence'][i]
print('[원문]:', sentence)

# 2. word 추출
start_idx = train_ver2['subject_entity'][i].find('start_idx')
start_idx2 = train_ver2['object_entity'][i].find('start_idx')
word1 = train_ver2['subject_entity'][i][10: start_idx-4]
word2 = train_ver2['object_entity'][i][10: start_idx2-4]
print('[word1]:',word1,'[word2]:', word2)

#3. word 교체
new_sentence = sentence.replace(word1, '🅰').replace(word2, '🅱').replace('🅰', word2).replace('🅱', word1)
print('[aug]:', new_sentence)
```

```python
'''
ex. id가 1인 문장
원문: 호남이 기반인 바른미래당·대안신당·민주평화당이 우여곡절 끝에 합당해 ... ( subject: 민주 평화당, object: 대안신당)
aug: 호남이 기반인 바른미래당·민주평화당·대안신당이 우여곡절 끝에 합당해 ... 
''' 

#i = 1
for i in idx:
    # 1. 원본 문장 확인
    sentence = train_ver2['sentence'][i]
    #print('[원문]:', sentence)

    # 2. word 추출
    start_idx = train_ver2['subject_entity'][i].find('start_idx')
    start_idx2 = train_ver2['object_entity'][i].find('start_idx')
    word1 = train_ver2['subject_entity'][i][10: start_idx-4]
    word2 = train_ver2['object_entity'][i][10: start_idx2-4]
    #print('[word1]:',word1,'[word2]:', word2)

    #3. word 교체
    new_sentence = sentence.replace(word1, '🅰').replace(word2, '🅱').replace('🅰', word2).replace('🅱', word1)
    #print('[aug]:', sentence)
    
    train_ver2['sentence'][i] = new_sentence

train_ver2
```
